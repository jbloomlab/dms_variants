```python
import glob
import os
import sys

from IPython.display import display, Image


from pymol import cmd

print('here')


epitope_colors = {'class 1': (0.12156862745098039, 0.4666666666666667, 0.7058823529411765)}
for f in glob.glob('*.pdb'):
    png = os.path.splitext(f)[0] + '.png'
    epitope = os.path.splitext(f)[0].split('-')[-1].replace('_', ' ')
    if epitope != 'class 1':
        continue
    cmd.reinitialize()
    cmd.load(f)
    cmd.bg_color('white')
    cmd.hide('everything')
    cmd.show('surface', 'chain E')
    cmd.set_color('max_color', epitope_colors[epitope])
    cmd.spectrum('b', f"white max_color")
    cmd.show('cartoon', '(chain A) and resi 18-85')
    cmd.set('cartoon_transparency', 0.6, 'chain A')
    cmd.color('gray', 'chain A')
    cmd.png(png, ray=1)
    display(Image(png))

    
```
